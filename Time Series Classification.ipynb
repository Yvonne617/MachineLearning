```python
clms=['min1','max1','mean1','median1','sd1','1stquart1','3rdquart1','min2','max2','mean2','median2','sd2','1stquart2','3rdquart2','min3','max3','mean3','median3','sd3','1stquart3','3rdquart3','min4','max4','mean4','median4','sd4','1stquart4','3rdquart4','min5','max5','mean5','median5','sd5','1stquart5','3rdquart5','min6','max6','mean6','median6','sd6','1stquart6','3rdquart6']

def timeSplit(n):
    n_para = 6
    n_fet = 7
    minn = [ [] for i in range(n_para*n)]
    maxn = [ [] for i in range(n_para*n)]
    mean = [ [] for i in range(n_para*n)]
    median = [ [] for i in range(n_para*n)]
    f_quartile = [ [] for i in range(n_para*n)]
    t_quartile = [ [] for i in range(n_para*n)]
    sd = [ [] for i in range(n_para*n)]

    for i in range(0,88,1):
        for nn in range(n):                                   #seperate into n part
            for x in range(n_para):                           # in the nn part  
                leng = len(data[i].iloc[:])
                a = int(leng/n)
                minn[x+nn*n_para].append(np.min(data[i].iloc[nn*a:(nn+1)*a,x+1]))
                maxn[x+nn*n_para].append(np.max(data[i].iloc[nn*a:(nn+1)*a,x+1]))
                mean[x+nn*n_para].append(np.mean(data[i].iloc[nn*a:(nn+1)*a,x+1]))
                median[x+nn*n_para].append(np.median(data[i].iloc[nn*a:(nn+1)*a,x+1]))
                sd[x+nn*n_para].append(np.std(data[i].iloc[nn*a:(nn+1)*a,x+1]))
                f_quartile[x+nn*n_para].append(np.percentile(data[i].iloc[nn*a:(nn+1)*a,x+1],25))
                t_quartile[x+nn*n_para].append(np.percentile(data[i].iloc[nn*a:(nn+1)*a,x+1],75))

    #return minn
    a = []
    a.append([minn,maxn,mean,median,sd,f_quartile,t_quartile])
    Df = pd.DataFrame()


    for nn in range(n):
        t_data={}
        for i in range(n_para):
            for j in range(n_fet):
                t_data[clms[7*i+j]+','+str(nn+1)+'slice'] = a[0][j][i+6*nn]
        #print(t_data)
        g=pd.DataFrame(t_data)
        Df = pd.concat([Df,g],axis=1)
    Df['move'] = df1['move']
    return Df
```

```python
def splitTestTrain(dataf):
    test=[]
    train=[]

    for i in range(0,88,1):
        if(i==0 or i==1 or i==7 or i==8 or i==13 or i==14 or i==15 or i==28 or i==29 or i==30 or i== 43 or i==44 or i==45 or i==58 or i==59 or i==60 or i==73 or i==74 or i==75):
            test.append(dataf.iloc[i])
        else:
            train.append(dataf.iloc[i])
        
    test = pd.DataFrame(test)
    test = test.reset_index(drop=True)
    train = pd.DataFrame(train)
    train = train.reset_index(drop=True)
    return train,test

def replaceLabel(dataf):
    dataf['move']=dataf['move'].replace({'bending1':1,'bending2':1})
    dataf['move']=dataf['move'].replace({'cycling':0,'lying':0,'sitting':0,'standing':0,'walking':0})
    return dataf    
```

```python

from sklearn.feature_selection import RFECV
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_predict
```

```python
scores=[]
chosenFeaturesbest=[ [[]for j in range(6)] for i in range(21)]

for i in range(1,21,1):
    train1,test1=splitTestTrain(timeSplit(i))
    train1 = replaceLabel(train1)
    x = train1.iloc[:,:-1]
    y = train1.iloc[:,-1]
    skf = StratifiedKFold(n_splits=5)
    j = 0
    s=0
    ma=0
    for train_index, test_index in skf.split(x, y):
        j = j+1
        estimator = LogisticRegression(C=10000)
        selector = RFECV(estimator,step=1,cv=StratifiedKFold(5),
              scoring='accuracy')
        selector.fit(x.loc[train_index], y.loc[train_index])
        s+=selector.score(x.loc[test_index], y.loc[test_index])
        ma+=max(selector.grid_scores_)
        for m in range(0,len(selector.ranking_),1):
            if selector.ranking_[m]==1:
                chosenFeaturesbest[i][j].append(x.columns.values.tolist()[m])
    print("when L= %d Optimal number of features : %d" %(i,selector.n_features_))
    print("The average score in the validation data we use:",s/5)
    print('The max scores in the model when we use optimal features',ma/5)
    print("***********")
    
```

```python
from sklearn.utils import resample

train_up,test_up=splitTestTrain(timeSplit(1))
train_up = replaceLabel(train_up)
test_up = replaceLabel(test_up)
df_majority = train_up[train_up.move==0]
df_minority = train_up[train_up.move==1]
df_minority_upsampled = resample(df_minority, 
                                 replace=True,     # sample with replacement
                                 n_samples=60,    # to match majority class
                                 random_state=123) # reproducible results

# Combine majority class with upsampled minority class
df_upsampled = pd.concat([df_majority, df_minority_upsampled])
x = pd.DataFrame(df_upsampled,columns = chosenFeaturesbest[1][4])
y = df_upsampled[['move']]
logreg = LogisticRegression(C=10000)
logreg.fit(x, y.values.ravel())
y_pretrain = logreg.predict(x)
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y, y_pretrain)
print(confusion_matrix)
```

```python
def upSample(trainm):
    df_majority1 = trainm[trainm.move=='cycling']
    df_majority2 = trainm[trainm.move=='lying']
    df_majority3 = trainm[trainm.move=='sitting']
    df_majority4 = trainm[trainm.move=='standing']
    df_majority5 = trainm[trainm.move=='walking']
    df_majority = pd.concat([df_majority1, df_majority2,df_majority3,df_majority4,df_majority5])
    df_minority1 = trainm[trainm.move=='bending1']
    df_minority2 = trainm[trainm.move=='bending2']
    df_minority_upsampled1 = resample(df_minority1, 
                                 replace=True,     # sample with replacement
                                 n_samples=10,    # to match majority class
                                 random_state=123) # reproducible results
    df_minority_upsampled2 = resample(df_minority2, 
                                 replace=True,     # sample with replacement
                                 n_samples=10,    # to match majority class
                                 random_state=123) # reproducible results
    df_upsampled1 = pd.concat([df_majority, df_minority_upsampled1])
    df_upsampled2 = pd.concat([df_upsampled1, df_minority_upsampled2])
    df_upsampled2 = df_upsampled2.reset_index(drop=True)
    return df_upsampled2
```

```python
#Use upsanmpling to do the multiclass classfication
scores4=[]
chosenFeatures4=[[[]for j in range(6)] for i in range(21)]

for i in range(1,21,1):
    df_m = timeSplit(i)
    trainm,testm=splitTestTrain(df_m)
    df_upsampled2 = upSample(trainm)
    x = df_upsampled2.iloc[:,:-1]
    y = df_upsampled2.iloc[:,-1]
    skf = StratifiedKFold(n_splits=5)
    j=0
    s=0
    ma=0
    for train_index, test_index in skf.split(x, y):
        j = j+1
        estimator = linear_model.LogisticRegression(penalty='l1',multi_class='ovr', solver='liblinear')
        selector = RFECV(estimator,step=1,cv=StratifiedKFold(5),
              scoring='accuracy')
        selector.fit(x.loc[train_index], y.loc[train_index].values.ravel())
        s+=selector.score(x.loc[test_index], y.loc[test_index])
        ma+=max(selector.grid_scores_)
        for m in range(0,len(selector.ranking_),1):
            if selector.ranking_[m]==1:
                chosenFeatures4[i][j].append(x.columns.values.tolist()[m])
    print("when L= %d:" %i)
    print("The average score in the validation data we use:",s/5)
    print('The max scores in the model when we use optimal features',ma/5)
    print("*************")
```
